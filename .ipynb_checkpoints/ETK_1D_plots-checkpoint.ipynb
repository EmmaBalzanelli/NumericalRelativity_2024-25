```python
import matplotlib.pyplot as plt
import numpy as np
from kuibit.simdir import SimDir
from kuibit.grid_data import UniformGrid
```

## Exact Solution

```python
#initial data from the Exact Riemann solver of https://www.brunogiacomazzo.org/?page_id=395
exact = np.loadtxt("Sod_exact.dat")
```

```python
x_exact=exact[:,0] #position
rho_exact=exact[:,1] #rest mass density
press_exact=exact[:,2] #note: this is pgas+pmag
#velocity
vx_exact=exact[:,3]
vy_exact=exact[:,4]
vz_exact=exact[:,5]
#magnetic field
By_exact=exact[:,6]
Bz_exact=exact[:,7]
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x_exact, rho_exact, label="exact")
ax.set_xlabel('x')
ax.set_ylabel(r'$\rho$')
ax.set_xlim(-0.5, 0.5)
plt.legend()

x_positions=[0.45,0.72,0.82]
y_positions=[0.65,0.40,0.22]
Title = ['rarefaction \n wave','contact \n discontinuity', 'shock \n wave']
[fig.text(
        x_positions[i],      # x-position (center of the figure)
        y_positions[i],     # y-position
        Title[i],           # The title text
        ha='center',        # Horizontal alignment: center
        va='top',           # Vertical alignment: top
        fontsize=11,) for i in range(3)]

fig.suptitle('Exact Sod solution for the Riemann Problem', fontsize=16)

plt.show()
#you'll see rarefaction wave and shocks waves
```

## Read the Numerical Results

```python
gf= SimDir("/home/balza/simulations/Sod400").gf #dx=0.0025
gf2= SimDir("/home/balza/simulations/Sod200").gf #dx=0.005
gf3= SimDir("/home/balza/simulations/Sod2000").gf #dx=0.0005
```

```python
vars1d=gf.x
vars1d2=gf2.x
vars1d3=gf3.x
print(vars1d)
```

```python
rho=vars1d.fields.rho
rho2=vars1d2.fields.rho
rho3=vars1d3.fields.rho

press=vars1d.fields.press
press2=vars1d2.fields.press
press3=vars1d3.fields.press

vx=vars1d.fields.vel[0]
vx2=vars1d2.fields.vel[0]
vx3=vars1d3.fields.vel[0]
```

## Kuibit Standard Methods for Plots

We create a uniform grid and plot the data on it. Kuibit indeed assumes that data were produced on a grid with several meshes (AMR). In our case the grid was already uniform, so later we will see an alternative way to plot the data without interpolating them on a new grid.

```python
small_grid = UniformGrid([101], x0=[-0.45], x1=[0.45])
```

### Rest-mass density

```python
print(max(rho.iterations))
print(max(rho2.iterations))
print(max(rho3.iterations))
```

```python
rho_final = rho.read_on_grid(max(rho.iterations), small_grid)
rho2_final = rho2.read_on_grid(max(rho2.iterations), small_grid)
rho3_final = rho3.read_on_grid(max(rho3.iterations), small_grid)
```

```python
#print(*rho_final.coordinates_meshgrid())
#print(*rho2_final.coordinates_meshgrid())
#print(*rho3_final.coordinates_meshgrid())
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x_exact, rho_exact, label="exact", color='black')
ax.plot(*rho2_final.coordinates_meshgrid(), rho2_final.data_xyz, '--', linewidth=1.5, label="GRHydro 200")
ax.plot(*rho_final.coordinates_meshgrid(), rho_final.data_xyz, '--', linewidth=1.5, label="GRHydro 400")
ax.plot(*rho3_final.coordinates_meshgrid(), rho3_final.data_xyz, '--', linewidth=1.5, label="GRHydro 2000")

ax.set_xlabel('x')
ax.set_ylabel(r'$\rho$')
ax.set_xlim(-0.5, 0.5)

x_positions=[0.45,0.72,0.82]
y_positions=[0.65,0.40,0.22]
Title = ['rarefaction \n wave','contact \n discontinuity', 'shock \n wave']
[fig.text(
        x_positions[i],      # x-position (center of the figure)
        y_positions[i],     # y-position
        Title[i],           # The title text
        ha='center',        # Horizontal alignment: center
        va='top',           # Vertical alignment: top
        fontsize=11,) for i in range(3)]

fig.suptitle('Sod rest-mass density final solutions', fontsize=16)
plt.legend()
plt.grid(True)

plt.savefig('./SODsol.png')

plt.show()
```

### Pressure

```python
press_final = press.read_on_grid(max(press.iterations), small_grid)
press2_final = press2.read_on_grid(max(press2.iterations), small_grid)
press3_final = press3.read_on_grid(max(press3.iterations), small_grid)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x_exact, press_exact, label="exact", color='black')
ax.plot(*press2_final.coordinates_meshgrid(), press2_final.data_xyz, '--', linewidth=1.5, label="GRHydro 200")
ax.plot(*press_final.coordinates_meshgrid(), press_final.data_xyz, '--', linewidth=1.5, label="GRHydro 400")
ax.plot(*press3_final.coordinates_meshgrid(), press3_final.data_xyz, '--', linewidth=1.5, label="GRHydro 2000")

ax.set_xlabel('x')
ax.set_ylabel(r'$P=P_{gas}+P_{mag}$')
ax.set_xlim(-0.5, 0.5)

x_positions=[0.45,0.82]
y_positions=[0.65,0.26]
Title = ['rarefaction \n wave', 'shock \n wave']
[fig.text(
        x_positions[i],      # x-position (center of the figure)
        y_positions[i],     # y-position
        Title[i],           # The title text
        ha='center',        # Horizontal alignment: center
        va='top',           # Vertical alignment: top
        fontsize=11,) for i in range(len(Title))]

fig.suptitle('Sod pression final solutions', fontsize=16)
plt.legend()
plt.grid(True)

plt.savefig('./SODsol_press.png')

plt.show()
```

### Velocity along x

```python
vx_final = vx.read_on_grid(max(vx.iterations), small_grid)
vx2_final = vx2.read_on_grid(max(vx2.iterations), small_grid)
vx3_final = vx3.read_on_grid(max(vx3.iterations), small_grid)
```

```python
fig, ax = plt.subplots(figsize=(7,5))
ax.plot(x_exact, vx_exact, label="exact", color='black')
ax.plot(*vx2_final.coordinates_meshgrid(), vx2_final.data_xyz, '--', linewidth=1.5, label="GRHydro 200")
ax.plot(*vx_final.coordinates_meshgrid(), vx_final.data_xyz, '--', linewidth=1.5, label="GRHydro 400")
ax.plot(*vx3_final.coordinates_meshgrid(), vx3_final.data_xyz, '--', linewidth=1.5, label="GRHydro 2000")

ax.set_xlabel('x')
ax.set_ylabel(r'$v_x$')
ax.set_xlim(-0.5, 0.5)

x_positions=[0.25,0.84]
y_positions=[0.55,0.55]
Title = ['rarefaction \n wave', 'shock \n wave']
[fig.text(
        x_positions[i],      # x-position (center of the figure)
        y_positions[i],     # y-position
        Title[i],           # The title text
        ha='center',        # Horizontal alignment: center
        va='top',           # Vertical alignment: top
        fontsize=11,) for i in range(len(Title))]

fig.suptitle('Sod velocity final solutions', fontsize=16)
plt.legend()
plt.grid(True)

plt.savefig('./SODsol_velx.png')

plt.show()
```
